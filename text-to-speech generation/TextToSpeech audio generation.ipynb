```python
w2n.word_to_num('one hundred')
```

```python
from google.colab import files
import time

for number in numbers[90:100]:
    tts = gTTS(number) #Provide the string to convert to speech
    word_to_num = w2n.word_to_num(number)
    tts.save(f'GTTSaudio/{word_to_num}.wav') #save the string converted to speech as a .wav file
    sound_file = f'GTTSaudio/{word_to_num}.wav'
    files.download(sound_file) 
#Audio(sound_file, autoplay=True) 
```
